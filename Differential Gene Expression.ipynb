```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

import os
# Libraries for DGE Analysis
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
# Library for Functional Enrichment
from gprofiler import GProfiler

# --- FILE NAMES ---
COUNT_MATRIX_FILE = "gene_count_matrix_loaded.csv"
MANIFEST_FILE = "sample_manifest_final.csv"

# DeseqStats documentation specifies a list
COMPARISON = ["Condition", "EGCG_Reversal", "Senescent_Control"] # Numerator vs Denominator

def run_deseq_and_enrichment(count_file: str, manifest_file: str):
    """
    Performs differential gene expression analysis using pydeseq2 and
    runs functional enrichment on the significant genes.
    """

    # --- 1. LOAD DATA ---
    print(f"Loading count matrix from {count_file} and manifest from {manifest_file}...")
    try:
        # ====================================================================
        # --- VITAL FIX FOR 'duplicate labels' ERROR ---
        # Load counts *without* index first to check for duplicates
        temp_counts = pd.read_csv(count_file)
        # Get the name of the first column (gene names)
        gene_col = temp_counts.columns[0]

        # Check for and aggregate duplicates
        duplicates = temp_counts[gene_col].duplicated().sum()
        if duplicates > 0:
            print(f"WARNING: Found {duplicates} duplicate gene IDs. Aggregating counts by summing...")
            # Group by the gene name column and sum all other (numeric) columns
            counts_df = temp_counts.groupby(gene_col).sum()
            print(f"Original gene count: {len(temp_counts)}, Aggregated gene count: {len(counts_df)}")
        else:
            print("No duplicate gene IDs found. Setting index.")
            counts_df = temp_counts.set_index(gene_col)
        # ====================================================================

        meta_df = pd.read_csv(manifest_file, index_col='Sample_ID')

    except FileNotFoundError as e:
        print(f"FATAL ERROR: Required file not found. {e}")
        return
    except Exception as e:
        print(f"An error occurred during data loading: {e}")
        return


    # --- 2. CRITICAL DATA ALIGNMENT (Mapping short labels to GSM IDs) ---
    label_to_gsm_map = {
        "ST_1": "GSM8727691", "ST_2": "GSM8727690", "ST_5": "GSM8727689",
        "SEN_1": "GSM8727688", "SEN_2": "GSM8727687", "SEN_3": "GSM8727686",
        "NT_1": "GSM8727685", "NT_2": "GSM8727684", "NT_3": "GSM8727683"
    }

    counts_df = counts_df.rename(columns=label_to_gsm_map)
    shared_samples = list(meta_df.index.intersection(counts_df.columns))

    if not shared_samples:
        print("ERROR: No shared samples found after mapping. Check file contents and map.")
        return

    counts_sub_df = counts_df[shared_samples]
    meta_sub_df = meta_df.loc[shared_samples].sort_index()

    print(f"Data alignment complete. Analyzing {len(shared_samples)} samples...")
    # --- PASTE THIS IN A NEW CELL AT THE BOTTOM OF YOUR NOTEBOOK ---

    # 1. SETUP: Load and Align Data (Just like in your main function)
    print("Loading data for diagnostics...")
    counts_df = pd.read_csv("gene_count_matrix_loaded.csv", index_col=0)
    meta_df = pd.read_csv("sample_manifest_final.csv", index_col='Sample_ID')

    # Aggregate duplicates if they exist (Critical step from your pipeline)
    if counts_df.index.duplicated().any():
        print("Aggregating duplicate genes...")
        counts_df = counts_df.groupby(counts_df.index).sum()

    # Map the column names
    label_to_gsm_map = {
        "ST_1": "GSM8727691", "ST_2": "GSM8727690", "ST_5": "GSM8727689",
        "SEN_1": "GSM8727688", "SEN_2": "GSM8727687", "SEN_3": "GSM8727686",
        "NT_1": "GSM8727685", "NT_2": "GSM8727684", "NT_3": "GSM8727683"
    }
    counts_df = counts_df.rename(columns=label_to_gsm_map)

    # Filter to shared samples
    shared_samples = list(meta_df.index.intersection(counts_df.columns))
    counts_sub_df = counts_df[shared_samples]
    meta_sub_df = meta_df.loc[shared_samples].sort_index()

    # 2. PLOT: Correlation Heatmap
    print("Generating Correlation Heatmap...")
    log_counts = np.log2(counts_sub_df + 1)
    corr_matrix = log_counts.corr(method='pearson')

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=0.90, vmax=1.0)
    plt.title("Sample Correlation Matrix\n(Dark Red = Identical Samples)")
    plt.show()

    # 3. PLOT: PCA
    print("Generating PCA Plot...")
    pca = PCA(n_components=2)
    # Transpose so samples are rows
    principal_components = pca.fit_transform(log_counts.T)
    pc1_var = round(pca.explained_variance_ratio_[0] * 100, 1)
    pc2_var = round(pca.explained_variance_ratio_[1] * 100, 1)

    plt.figure(figsize=(9, 7))
    group_colors = {'Normal_Control': 'blue', 'Senescent_Control': 'red', 'EGCG_Reversal': 'green'}

    for i, sample_name in enumerate(log_counts.columns):
        condition = meta_sub_df.loc[sample_name, 'Condition']
        plt.scatter(principal_components[i, 0], principal_components[i, 1],
                    color=group_colors.get(condition, 'gray'), s=100, alpha=0.8,
                    label=condition)
        plt.text(principal_components[i, 0]+0.2, principal_components[i, 1]+0.2, sample_name, fontsize=8)

    # Clean up legend (remove duplicates)
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys())

    plt.xlabel(f'PC1 ({pc1_var}%)')
    plt.ylabel(f'PC2 ({pc2_var}%)')
    plt.title('PCA Plot: Visualizing Group Separation')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
    # --- POSITIVE CONTROL: Normal vs. Senescent ---
    print("--- RUNNING POSITIVE CONTROL: Normal_Control vs. Senescent_Control ---")

    # 1. Select Samples
    comparison_groups = ["Normal_Control", "Senescent_Control"]
    samples_pos_control = meta_sub_df[meta_sub_df['Condition'].isin(comparison_groups)].copy()
    counts_pos_control = counts_sub_df[samples_pos_control.index].astype(int)

    # 2. Pre-filter low counts (Crucial for p-value calculation)
    genes_to_keep = counts_pos_control.sum(axis=1) >= 10
    counts_pos_control = counts_pos_control[genes_to_keep]

    # 3. Create DESeq2 Object
    # Note: "Senescent_Control" is the reference (denominator), "Normal_Control" is the numerator
    dds_pos = DeseqDataSet(
        counts=counts_pos_control.T,
        metadata=samples_pos_control,
        design_factors="Condition"
    )

    # 4. Run Analysis
    dds_pos.deseq2()
    stat_res_pos = DeseqStats(dds_pos, contrast=["Condition", "Normal_Control", "Senescent_Control"])
    stat_res_pos.summary()

    # 5. Check Results
    res_pos = stat_res_pos.results_df
    sig_pos = res_pos[(res_pos['padj'] < 0.05) & (abs(res_pos['log2FoldChange']) > 0.5)]

    print(f"\n--- POSITIVE CONTROL RESULTS ---")
    print(f"Comparison: Normal vs. Senescent")
    print(f"Significant Genes Found: {len(sig_pos)}")

    if len(sig_pos) > 100:
        print("SUCCESS: Pipeline is working. The lack of EGCG signal is a biological result.")
    else:
        print("FAILURE: Something is wrong with the data loading or normalization.")

    # --- PCA PLOT GENERATION ---

    # 1. Log-Transform Data for PCA (Simple log2(x+1))
    # We use the full dataset (all 9 samples) to see where they all sit
    log_counts = np.log2(counts_sub_df + 1).T  # Transpose: Rows = Samples

    # 2. Run PCA
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(log_counts)
    pc1_var = round(pca.explained_variance_ratio_[0] * 100, 1)
    pc2_var = round(pca.explained_variance_ratio_[1] * 100, 1)

    # 3. Plot
    plt.figure(figsize=(10, 8))
    colors = {'Normal_Control': 'blue', 'Senescent_Control': 'red', 'EGCG_Reversal': 'green'}

    for i, sample in enumerate(log_counts.index):
        condition = meta_sub_df.loc[sample, 'Condition']
        plt.scatter(principal_components[i, 0], principal_components[i, 1],
                    c=colors[condition], label=condition if condition not in plt.gca().get_legend_handles_labels()[1] else "",
                    s=100, alpha=0.8)
        plt.text(principal_components[i, 0]+0.5, principal_components[i, 1]+0.5, sample, fontsize=9)

    plt.xlabel(f'PC1 ({pc1_var}%)')
    plt.ylabel(f'PC2 ({pc2_var}%)')
    plt.title('PCA Plot: Transcriptomic distinctness of groups')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    # --- 3. DIFFERENTIAL EXPRESSION ANALYSIS (pydeseq2) ---

    # Filter to only the samples involved in the primary comparison
    samples_for_dge = meta_sub_df[
        meta_sub_df['Condition'].isin([COMPARISON[1], COMPARISON[2]])
    ].copy()

    # Get the corresponding counts and ensure they are integers
    counts_for_dge = counts_sub_df[samples_for_dge.index].astype(int)

    # --- FIX FOR 'NaN' p-value ERROR ---
    # Pre-filter low-count genes. Keep genes with at least 10 total
    # reads across the 6 samples being analyzed.
    min_total_counts = 10
    genes_to_keep = counts_for_dge.sum(axis=1) >= min_total_counts
    counts_for_dge = counts_for_dge[genes_to_keep]

    print(f"Pre-filtering: Kept {len(counts_for_dge)} genes (out of {len(genes_to_keep)}) with >= {min_total_counts} total reads.")

    # --- FIX FOR 'ValueError: ... ambiguous' ---
    # Convert to category dtype
    samples_for_dge['Condition'] = samples_for_dge['Condition'].astype('category')

    # --- FIX FOR 'ValueError: Design matrix has an empty column' ---
    # Remove the 'Normal_Control' level, which is no longer present
    samples_for_dge['Condition'] = samples_for_dge['Condition'].cat.remove_unused_categories()

    print(f"\n--- Running DESeq2 Analysis on {len(samples_for_dge)} samples ({COMPARISON[1]} vs {COMPARISON[2]}) ---")

    # Sanity check to ensure alignment
    if not counts_for_dge.T.index.equals(samples_for_dge.index):
        print("FATAL ERROR: Indices of counts and metadata do not match.")
        return

    # Create DeseqDataSet object
    # 'counts' expects samples as rows, genes as columns
    # 'clinical' (metadata) expects samples as rows
    dds = DeseqDataSet(
        counts=counts_for_dge.T,
        metadata=samples_for_dge,
        design_factors="Condition"
    )

    # Run DESeq analysis
    # This line should no longer fail due to duplicate labels
    dds.deseq2()

    # Extract results for the specified contrast
    stat_res = DeseqStats(dds, contrast=COMPARISON)

    # Run the statistical analysis
    stat_res.summary()
    results_df = stat_res.results_df

    # Identify significant genes
    results_df['significant'] = (results_df['padj'] < 0.05) & (abs(results_df['log2FoldChange']) > 0.5)

    significant_genes = results_df[results_df['significant']].sort_values(
        'log2FoldChange', ascending=False
    ).dropna(subset=['padj'])

    results_filename = "DGE_Results_EGCG_vs_Senescent.csv"
    results_df.to_csv(results_filename)

    print(f"\n DGE Analysis Complete: {results_filename} saved.")
    print(f"   Total genes analyzed: {len(results_df)}")
    print(f"   Statistically significant genes (padj < 0.05 & |log2FC| > 1): {len(significant_genes)}")

    # --- 4. FUNCTIONAL ENRICHMENT ANALYSIS (gprofiler-official) ---

    if len(significant_genes) > 0:
        print("\n--- Running Functional Enrichment (GO, KEGG) on Significant Genes ---")

        # Separate up and down regulated genes
        up_genes = significant_genes[significant_genes['log2FoldChange'] > 0].index.tolist()
        down_genes = significant_genes[significant_genes['log2FoldChange'] < 0].index.tolist()

        gp = GProfiler(return_dataframe=True)

        # --- FIX FOR 'ValueError: All gene IDs are 'nan'' ---
        # Check if lists are empty before running enrichment
        if up_genes:
            print(f"   Processing {len(up_genes)} up-regulated genes...")
            enrichment_up = gp.profile(organism='hsapiens', query=up_genes,
                                       no_iea=True, ordered=False,
                                       significance_threshold_method='fdr')
            enrichment_up.to_csv("Enrichment_UpRegulated.csv", index=False)
            print(f"   Enrichment for up-regulated genes saved to 'Enrichment_UpRegulated.csv'.")
        else:
            print("   No significant UP-regulated genes found. Skipping enrichment.")

        if down_genes:
            print(f"   Processing {len(down_genes)} down-regulated genes...")
            enrichment_down = gp.profile(organism='hsapiens', query=down_genes,
                                         no_iea=True, ordered=False,
                                         significance_threshold_method='fdr')
            enrichment_down.to_csv("Enrichment_DownRegulated.csv", index=False)
            print(f"   Enrichment for down-regulated genes saved to 'Enrichment_DownRegulated.csv'.")
        else:
            print("   No significant DOWN-regulated genes found. Skipping enrichment.")

    else:
        print("\nSkipping Functional Enrichment: No significant genes found.")


if __name__ == "__main__":
    # Ensure pydeseq2 is installed: pip install pydeseq2
    # Ensure gprofiler-official is installed: pip install gprofiler-official
    run_deseq_and_enrichment(COUNT_MATRIX_FILE, MANIFEST_FILE)
```
